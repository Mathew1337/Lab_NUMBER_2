```
import base64
import json

class BusStop:
    def __init__(self, name, lat, lon, time_to_next):
        self.name = name
        self.lat = lat
        self.lon = lon
        self.time_to_next = time_to_next

    def to_base64(self):
        return base64.b64encode(json.dumps(self.__dict__).encode()).decode()

    @classmethod
    def from_base64(cls, s):
        return cls(**json.loads(base64.b64decode(s).decode()))


class BusRoute:
    def __init__(self):
        self.stops = []

    def add(self, name, lat, lon, time_to_next):
        self.stops.append(BusStop(name, lat, lon, time_to_next))

    def total_time(self):
        return sum(s.time_to_next for s in self.stops[:-1])

    def where(self, n):
        if not self.stops:
            return "Нет остановок"
        return self.stops[n % len(self.stops)].name

    def reverse(self):
        r = BusRoute()
        for i in range(len(self.stops)-1, -1, -1):
            t = self.stops[i-1].time_to_next if i > 0 else 0
            r.add(self.stops[i].name, self.stops[i].lat, self.stops[i].lon, t)
        return r

    def save_base64(self, file):
        with open(file, 'w') as f:
            for s in self.stops:
                f.write(s.to_base64() + '\n')

    def load_base64(self, file):
        self.stops = []
        with open(file, 'r') as f:
            for line in f:
                if line.strip():
                    self.stops.append(BusStop.from_base64(line.strip()))


route = BusRoute()
route.add("Бауманская", 55.772, 37.679, 4)
route.add("Электрозаводская", 55.772, 37.703, 5)
route.add("Сокольники", 55.789, 37.680, 6)
route.add("Китай-город", 55.755, 37.633, 7)
route.add("Курская", 55.757, 37.658, 0)

print(f"Общее время: {route.total_time()} мин")
print(f"Через 3 остановки: {route.where(3)}")
print(f"Через 7 остановок: {route.where(7)}")

rev = route.reverse()
print("Обратный маршрут:")
for i, s in enumerate(rev.stops, 1):
    print(f"{i}. {s.name} ({s.time_to_next} мин до след)")

route.save_base64("route.txt")
print("Сохранено в route.txt")
```
